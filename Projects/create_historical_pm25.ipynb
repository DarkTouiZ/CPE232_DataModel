```python
import pandas as pd
```

```python
df = pd.read_csv('air_quality_northern.csv', index_col=0, parse_dates=True)
df = df.rename(columns={'วันที่': 'date'})
df['date'] = pd.to_datetime(df['date'])
```

```python
daily_mean = df.groupby(df['date'].dt.date)[['73t', '41t', '75t', '58t', '39t', '76t']].mean()
daily_mean
```

```python
daily_mean.to_csv('dataset/PM2.5/1m_history_PM25.csv', index=True)
```
